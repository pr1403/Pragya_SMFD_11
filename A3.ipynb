# Q2

```python
import random
import numpy as np
import matplotlib.pyplot as plt

T = 100
alpha = 0.05
sigma = 0.2
S=[]
for j in range(1,6):
    s1 = [1]
    w = [0]
    t1 = [0]
    for i in range(T):
        w.append(w[-1]+random.gauss(0, 1))
        tc=i+1
        exp=sigma*w[-1]+(alpha-0.5*sigma**2)*tc
        s1.append(s1[0]*np.exp(exp))
        t1.append(tc)
    S.append(s1)

plt.plot(t1,S[0],color='red')
plt.plot(t1,S[1],color='blue')
plt.plot(t1,S[2],color='green')
plt.plot(t1,S[3],color='yellow')
plt.plot(t1,S[4],color='orange')
```

# Q3

```python
T=1000
paths=10000
s=500
t=700
W=np.cumsum(np.random.randn(paths,T),axis=1)
Ws=W[:,s+1]
Wt=W[:,t+1]
e_ws_wt=np.mean(Ws*Wt)
e_ws_wt
```

# Q4

```python
diff=Ws-Wt
mu=np.mean(diff)
var=np.var(diff)
plt.hist(diff,bins=25)
print(f'Mean={mu}, Variance={var}')
```

```python
#independence
from scipy.stats import pearsonr
v=100
u=400
A = W[:, t] - W[:, s]
B = W[:, v] - W[:, u]
corr,etc=pearsonr(A,B)
corr
```


# Q5

```python
coeffs = np.polyfit(Ws, Wt, 1)
slope, intercept = coeffs

print(f"E[Wt | Ws] ≈ {slope} * Ws + {intercept}")

# Plot
plt.scatter(Ws, Wt, s=1, alpha=0.05)
x = np.linspace(min(Ws), max(Ws), 100)
plt.plot(x, slope * x + intercept)
plt.legend()
plt.grid(True)
```
